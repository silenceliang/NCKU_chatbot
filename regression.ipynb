```python
''' import external modules '''
import numpy as np
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
''' import iris datasets '''
iris = datasets.load_iris()

X = iris.data  
y = iris.target
```

```python
X_b = np.concatenate((X, np.ones(len(X)).reshape(-1,1)), axis=1)
X_b # ax + bias = y
```

```python
len(y),y
```

```python
def least_square_equation(X, y):
    # (X.T*X)^-1*X.T*y
    return np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
```

```python
def sum_of_square_error(fx, y):
    E = y - fx
    # E.T*E
    SSE = np.dot(E.T,E)
    return SSE
```

```python
def linear_regression(X, y):
    W = least_square_equation(X, y)
    return W
```

```python
weight = linear_regression(X_b,y) # get weight
f = np.sum(weight*X_b, axis=1) # weight * X
SSE = sum_of_square_error(f, y) # get current error
```
